```python
# Importing the pandas library and aliasing it as 'pd' for easier reference
import pandas as pd
```

```python
# Reading the 'Salary_Data.csv' file into a DataFrame named 'dataset'
dataset=pd.read_csv("Salary_Data.csv")
```

```python
dataset
```

```python
independent=dataset[["YearsExperience"]]
# Selecting the 'YearsExperience' column from the 'dataset' DataFrame and storing it in the 'independent' variable
```

```python
independent
```

```python
dependent=dataset[["Salary"]]
# Selecting the 'Salary' column from the 'dataset' DataFrame and storing it in the 'dependent' variable
```

```python
dependent
```

```python
# Importing the 'train_test_split' function from the 'sklearn.model_selection' module
from sklearn.model_selection import train_test_split
# Splitting the 'independent' and 'dependent' variables into training and testing sets
# 30% of the data is reserved for testing, and a random state of 0 is set for reproducibility
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0) 
```

```python
# Importing the 'LinearRegression' class from the 'sklearn.linear_model' module
from sklearn.linear_model import LinearRegression
# Creating an instance of the 'LinearRegression' class and storing it in the 'regressor' variable
regressor=LinearRegression()
# Training the linear regression model using the training data
regressor.fit(X_train,y_train)
```

```python
# Retrieving the coefficients (weights) of the linear regression model
weight=regressor.coef_
weight
```

```python
# Retrieving the intercept (bias) of the linear regression model
bais=regressor.intercept_
bais
```

```python
# Making predictions on the test set using the trained linear regression model
y_pred=regressor.predict(X_test)
```

```python
# Importing the 'r2_score' function from the 'sklearn.metrics' module to evaluate model performance
from sklearn.metrics import r2_score
# Calculating the R-squared score to evaluate how well the model's predictions match the actual values
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
# Importing the pickle library to save and load models
import pickle
# Defining the filename for saving the model
filename="finilized_model_linear.sav"
```

```python
# Saving the trained model to a file
pickle.dump(regressor,open(filename,'wb'))
```

```python
# Loading the saved model from the file
loaded_model=pickle.load(open("finilized_model_linear.sav",'rb'))
# Making a prediction using the loaded model for 13 years of experience
result=loaded_model.predict([[13]])
```
